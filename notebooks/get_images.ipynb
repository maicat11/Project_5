## Rename and move files into train/valid/test splits

```python
import os 
  
# Function to rename multiple files 
def rename_file(directory, prefix): 
    for i, filename in enumerate(os.listdir(directory)): 
        new_name = prefix + str(i) + ".jpg"
        source = directory + filename 
        dir_and_name = directory + new_name 
        os.rename(source, dir_and_name) 
    print("Done!")
```

```python
# rename the files used for creating pipeline
img_dir = '../data/google_api/'
items = ['black_dresses', 'black_pants', 'black_shorts', 
         'blue_dresses', 'evening_dresses', 'red_dresses', 
         'spring_dresses']
for item in items:
    path = img_dir + item + '/'
    rename_file(path, item)
```

```python
import random
from pathlib import Path

def move_files(dir_name):
    """
    Move images into train, validate, test directories.
    Ensure the distribution is random while moving files.
    Train, validate, test is 0.8/0.1/0.1 split.
    """
    current_path = img_dir + dir_name + '/'  
    train_path = '../data/clothing/train/' + dir_name + '/'
    val_path = '../data/clothing/validate/' + dir_name + '/'
    test_path = '../data/clothing/test/' + dir_name + '/'

    num_train = int(len(os.listdir(current_path))*0.8)  
    num_test = int(len(os.listdir(current_path))*0.1)  

    for _ in range(num_train):
        try:
            rand_file = random.choice(os.listdir(current_path))
            Path(current_path + rand_file).rename(train_path + rand_file)
        except FileNotFoundError:
            continue

    for _ in range(num_test):
        try:
            rand_file = random.choice(os.listdir(current_path))
            Path(current_path + rand_file).rename(val_path + rand_file)
        except FileNotFoundError:
            continue

    for _ in range(num_test):
        try:
            rand_file = random.choice(os.listdir(current_path))
            Path(current_path + rand_file).rename(test_path + rand_file)
        except FileNotFoundError:
            continue
    
```

```python
directories = ['black_pants', 'black_shorts', 'blue_dresses', 'red_dresses']

for d in directories:
    move_files(d)
```

# Shoes cleaning

```python
import pandas as pd
```

```python
shoe_info = pd.read_csv('../data/ut-zap50k-data/meta-data.csv')
print(shoe_info.shape)
display(shoe_info.head())
```

```python
def remove_nonwomen_shoe(full_path, file_name):
    cid = file_name.replace('.', '-')[:-4]
    try:
        row = shoe_info[shoe_info['CID'] == cid]
        if row['Gender'][row['Gender'].index[0]] != 'Women':
#             print(row['Gender'][row['Gender'].index[0]]) 
            os.remove(full_path + file_name)
            print(f'Removed {file_name}')
    except IndexError:
        print(f'Error at {cid}')        
```

```python
root = '../data/ut-zap50k-images/'
folders = ['ankle_boots/', 'athletic/', 'backless/', 'calf_boots/', 
           'flats/', 'heels/', 'knee_boots/', 'sandals/', 'slippers/']

for folder in folders:
    for file in os.listdir(root+folder):
#         print(root+folder+file)
        remove_nonwomen_shoe(root+folder, file)
    
```
